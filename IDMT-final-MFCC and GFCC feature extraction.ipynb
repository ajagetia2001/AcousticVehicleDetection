```python
#good_truck_kmeans.txt contains name of good truck samples based on RSME K-Means

with open('good_truck_kmeans.txt') as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]
train_truck=[]
for string in lines:
    string= "C:\\Users\\Mohd Ashhad\\Downloads\\IDMT_Traffic\\IDMT_Traffic\\audio\\"+string+".wav"
    train_truck.append(string)    
```

```python
#Function to implement time stretching

def time_stretch(signal,stretch_rate):
     return librosa.effects.time_stretch(signal,stretch_rate)
```

```python
# Fucntion that calls original IDMT audios, augments them and saved the augmented files in the origin folder with a new name

train_new=[]
for aud in train_truck:
    
    if '-BG' in aud:
        train_new.append(aud)
    else:
#         train_new.append(aud)
        required_audio_size = 2
        signal,sr=librosa.load(aud)
        augmented_signal=time_stretch(signal,1.5)
        padded_signal = fix_length(augmented_signal, size=required_audio_size*sr) 
        parts = aud.split('_')
        parts[5]=parts[5]+"aug1"
        name="_".join(parts)

        sf.write(file=name,data=padded_signal,samplerate=sr)
    
        train_new.append(name)
```

```python

train_new1=[]
for aud in train_truck:
    
    if '-BG' in aud:
        train_new1.append(aud)
    else:
#         train_new.append(aud)
        required_audio_size = 2
        signal,sr=librosa.load(aud)
        augmented_signal=time_stretch(signal,0.9)
        padded_signal = fix_length(augmented_signal, size=required_audio_size*sr) 
        parts = aud.split('_')
        parts[5]=parts[5]+"aug2"
        name="_".join(parts)
#         print(name)
        sf.write(file=name,data=padded_signal,samplerate=sr)
    
        train_new1.append(name)
```

```python

train_new2=[]
for aud in train_truck:
    
    if '-BG' in aud:
        train_new2.append(aud)
    else:
#         train_new.append(aud)
        required_audio_size = 2
        signal,sr=librosa.load(aud)
        augmented_signal=time_stretch(signal,1.2)
        padded_signal = fix_length(augmented_signal, size=required_audio_size*sr) 
        parts = aud.split('_')
        parts[5]=parts[5]+"aug3"
        name="_".join(parts)
#         print(name)
        sf.write(file=name,data=padded_signal,samplerate=sr)
    
        train_new2.append(name)
```

```python
all_trucks=train_truck+train_new+train_new1+train_new2
all_trucks
```

```python
def features_extractor2(file):
    
    # fucntion that extracts features
    
    y, sr = librosa.load(file)
    S = librosa.feature.mfcc(y, sr=sr, n_mels=128,n_fft=2048,hop_length=512,win_length=1024,n_mfcc=128)

    return S

# for GFCC use the following feature extraction function


# def features_extractor2(file):
    
#     import scipy
#     # !pip install spafe
#     from spafe.features.gfcc import gfcc
#     import scipy.io.wavfile
#     import spafe.utils.vis as vis
    
#     y, sr = librosa.load(file)
#     gfccs = gfcc(sig=y, fs=sr, num_ceps=64,nfilts=128)
#     return gfccs




def preproc_dataset(arr):
    
# function to build the pandas dataframe

    mfcc=[]
    for v in arr:
        mfcc_feature=features_extractor2(v)
        lis=mfcc_feature.tolist()
        v = v.replace('.wav', '')
        v=v[63 :]
        parts = v.split('_')

        # background noise files
        if '-BG' in v:
            date_time, location, speed_kmh, sample_pos, mic, channel = parts
            vehicle, source_direction, weather, daytime = 'None', 'None', 'None', 'None'
            is_background = True

        # files with vehicle passings
        else:
            date_time, location, speed_kmh, sample_pos, daytime, weather, vehicle_direction, mic, channel = parts
            vehicle, source_direction = vehicle_direction
            is_background = False

        channel = channel.replace('-BG', '')
        speed_kmh = speed_kmh.replace('unknownKmh', 'UNK')
        speed_kmh = speed_kmh.replace('Kmh', '')

        mfcc.append({'file': v,
                           'is_background': is_background,
                           'date_time': date_time,
                           'location': location,
                           'speed_kmh': speed_kmh,
                           'sample_pos': sample_pos,
                           'daytime': daytime,
                           'weather': weather,
                           'vehicle': vehicle,
                           'source_direction': source_direction,
                           'microphone': mic,
                           'channel': channel,
                           'mfcc_feature': lis})
          
        
    df_dataset = pd.DataFrame(mfcc, columns=('file', 'is_background', 'date_time', 'location', 'speed_kmh', 'sample_pos', 'daytime', 'weather', 'vehicle',
                                                       'source_direction', 'microphone', 'channel','mfcc_feature'))      
    return df_dataset
```
